# MNIST USPS Transport Distances

This notebook shows an example of computing optimal transport distances between the MNIST dataset and the USPS dataset with and without label distances

This notebook uses Pytorch to download the MNIST and USPS digit datasets. If necessary, run:

`conda install pytorch torchvision torchaudio -c pytorch`

```python
import sys
sys.path.append('../OTDD/')

import torch
from torchvision import datasets, transforms
import pandas as pd
import numpy as np

from otdd import *
from plot import *
from otdd_pytorch import *

from mnist_helper import *
```

# Load Data

```python
mnist_data = get_mnist_data()
```

```python
usps_data = get_usps_data()
```

```python
# subset to 5000 examples for speed
mnist_sample = mnist_data.subsample(5000, equal_classes=True)
usps_sample = usps_data.subsample(5000, equal_classes=True)
```

# Tensorized Distance and Cost Functions

First we'll use the tensorized Sinkhorn routines from the Geomloss Library

```python
distance_tensorized = PytorchEuclideanDistance()
routine_tensorized = SinkhornTensorized(distance_tensorized)
cost_tensorized = SamplesLossTensorized(routine_tensorized)
```

# Feature Distance

Here we compute the cost and optimal coulpling using only the feature space

```python
outputs = cost_tensorized.distance(mnist_sample.features, usps_sample.features, return_transport=True)
```

```python
outputs[0] # transportation cost
```

```python
plot_coupling(outputs[1].numpy(), outputs[2].numpy(), mnist_sample.labels.numpy(), usps_sample.labels.numpy(),
                  mnist_sample.classes, usps_sample.classes, figsize=(10,10), cmap='OrRd')
```

When we pass `return_transport=True`, we also get a series of vectors that encode the transport plan in linear memory

```python
F_i, G_j, a_i, b_j = outputs[-1]
```

```python
C_ij = outputs[1]
```

We can reconstruct the transpot plan (coupling matrix) with the formula:

\begin{align}
\pi ~=~ \exp\frac{1}{\epsilon}(f\oplus g-C)\,\cdot\,\alpha\otimes\beta
\end{align}

```python
P_ij = ((F_i + G_j - C_ij) / routine_tensorized.eps).exp() * (a_i * b_j)
```

# Label Distance

Here we compute the cost and optimal coupling with label distances

```python
outputs2 = cost_tensorized.distance_with_labels(mnist_sample.features, usps_sample.features, 
                                              mnist_sample.labels, usps_sample.labels, gaussian_class_distance=False)
```

```python
outputs2[0] # transport cost
```

Using label distances incorporates class-to-class similarity in coupling

```python
plot_coupling(outputs2[1].numpy(), 
              outputs2[2].numpy(), 
              mnist_sample.labels.numpy(), 
              usps_sample.labels.numpy(),
                  mnist_sample.classes, usps_sample.classes, figsize=(10,10), cmap='OrRd')
```

Here are the label-to-label distances

```python
plot_class_distances(outputs2[3], mnist_sample.classes, usps_sample.classes, 
                            cmap='OrRd', figsize=(10,8), text=True)
```

# Online Distance and Cost Functions

We can also use online routines for the Sinkhorn cost. These routines use KeOps LazyTensors to significantly reduce memory footprint. Around 5000 or more samples there starts to be a significant difference in memory usage

```python
distance_online = EuclideanOnline()
routine_tensorized = SinkhornOnline(distance_online)
cost_online = SamplesLossOnline(routine_tensorized)
```

```python
outputs = cost_online.distance(mnist_sample.features, usps_sample.features)
```
